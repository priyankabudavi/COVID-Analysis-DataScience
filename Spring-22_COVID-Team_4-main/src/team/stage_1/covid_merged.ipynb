## Merge COVID-19 datasets

## Here we are merging three covid datasets 

- covid_confirmed_usafacts.csv 
    Gives info about no. of covid confirmed cases 
- covid_deaths_usafacts.csv 
    Gives info about no. of covid death cases
         
- covid_county_population_usafacts.csv
    Gives info about population in county and state
        

## Importing libraries

```python
import pandas as pd
```

## Reading the covid19 dataset

```python
f1 = pd.read_csv('../../../data/stage_1/covid_confirmed_usafacts.csv')
f2 = pd.read_csv('../../../data/stage_1/covid_deaths_usafacts.csv')
f3 = pd.read_csv('../../../data/stage_1/covid_county_population_usafacts.csv')
```

### Here, we replace the space between the column header names with an underscore.

```python
f1.columns = f1.columns.str.replace(' ', '_')
f2.columns = f2.columns.str.replace(' ', '_')
f3.columns = f3.columns.str.replace(' ', '_')
```

### Use melt() function to convert the dataframe from wide to long format in confirmed cases and death cases .csv file

```python
m1 = pd.melt(f1,id_vars=['countyFIPS','County_Name','State','StateFIPS'],var_name = 'Date',value_name='Confirmed')
m1
```

```python
m2 = pd.melt(f2,id_vars=['countyFIPS','County_Name','State','StateFIPS'],var_name = 'Date',value_name='Deaths')
m2
```

#### Delete StateFIPS as population dataset don't have StateFIPS column. 

```python
del m1['StateFIPS']
del m2['StateFIPS']
```

#### Date is object datatype by default. So we use to_datetime() function to convert the Date to datetime format

```python
m1['Date']=pd.to_datetime(m1['Date'])
m2['Date']=pd.to_datetime(m2['Date'])
```

#### Merge confirmed cases and death cases covid files

```python
merge1 = pd.merge(m1,m2)
merge1
```

### Convert County_Name to lowercase and remove extra spaces so it can be merged easily

```python
merge1['County_Name'] = merge1['County_Name'].str.lower()
f3['County_Name'] = f3['County_Name'].str.lower()

merge1['County_Name'] = merge1['County_Name'].str.strip()
f3['County_Name'] = f3['County_Name'].str.strip()
```

### Finally, merging the population dataset 

```python
merge_final = pd.merge(merge1,f3)
merge_final
```

## Exporting the merged file to csv

```python
merge_final.to_csv('../../../data/stage_1/superdataset.csv',index = False)
```
