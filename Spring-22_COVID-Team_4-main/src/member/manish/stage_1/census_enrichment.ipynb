##### 1. Covid19 trends for the last week of the data.

Here I have chosen North Carolina state for showing covid trend

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
super_file = pd.read_csv('../../../../data/stage_1/superdataset.csv')
super_file
```

```python
nc_dataset = super_file[super_file['State'] == 'NC']
nc_dataset = nc_dataset[['State', 'County_Name', 'Date', 'Confirmed', 'Deaths', 'population']]
nc_dataset
```

#### We will try to show the trend in covid data using matplot.

#### Lets sum the confirmed cases after grouping by date and then take out last 7 days data using tail(7) function.

```python
total_confirmed = nc_dataset.groupby('Date')['Confirmed'].sum()
total_confirmed = total_confirmed.tail(7)
total_confirmed
```

#### Now, we sum the death cases after grouping by date and then taking out last 7 days data using tail(7) function.

```python
total_death = nc_dataset.groupby('Date')['Deaths'].sum()
total_death = total_death.tail(7)
total_death
```

## Confirmed cases graph

```python
confirmed_plot = total_confirmed.plot(title = 'Covid19 confirmed cases in North Carolina', legend=True, figsize=(8,5))
confirmed_plot.set_xlabel('Date')
confirmed_plot.set_ylabel('Confirmed Cases')
confirmed_plot
```

## Death cases graph

```python
death_plot = total_death.plot(title = 'Covid19 death cases in North Carolina', legend=True, figsize=(8,5))
death_plot.set_xlabel('Date')
death_plot.set_ylabel('Death cases')
death_plot
```

## Covid19 trend in North Carolina (Analysis):

##### 1. Confirmed cases:
- The graph of confirmed cases shows that the total no. of cases are increasing during the week.
- The flat line shows that no new cases are reported from 02/04/22 to 02/06/22 maybe due to weekend.

##### 2. Death cases:

- The flat line shows that no such cases are reported from 02/04/22 to 02/06/22 maybe due to weekend.
- The cases again starts increasing from Monday.

### There are few junk values in the dataset. So I used excel to remove junk values, which is '?' in my case.

### I have also removed first row which gives metadata information

```python
#Removing certain rows with junk values
census_file = census_file.drop([0])
census_file = census_file.drop([829])
census_file = census_file.drop([833])
census_file
```

### Now, we'll filter the countyFIPS from GEO_ID using split() function.

### And separate the county_name from State in NAME using split()

```python
#We will split the state name from county name and convert it to lower case.
#In GEO_ID, strip the front of the it to get countyFIPS

census_file['NAME'] = census_file['NAME'].apply(lambda p: p.strip())
census_file['NAME'] = census_file['NAME'].apply(lambda p: p.split(',')[0])
census_file['NAME'] = census_file['NAME'].apply(lambda cn: cn.lower())

census_file['GEO_ID'] = census_file['GEO_ID'].apply(lambda q: q.split('US')[1])
census_file['GEO_ID'] = census_file['GEO_ID'].apply(pd.to_numeric)
census_file
```

#### The column header as you can see currently are not providing much information about the column values. 
#### Let's rename the column name to make the dataset more understandable.
#### We will use pandas rename() function to rename the column headers. We will keep inplace =True, becuase we want the changes to be applied on the same dataframe.

```python
# changing the column headers.
census_file.rename(columns = {'NAME':'County_Name'}, inplace = True)

census_file.rename(columns = {'GEO_ID':'countyFIPS'}, inplace = True)

census_file.rename(columns = {'DP05_0001E':'Total_pop'}, inplace = True)

census_file.rename(columns = {'DP05_0002E':'Male_total_population'}, inplace = True)

census_file.rename(columns = {'DP05_0003E':'Female_total_population'}, inplace = True)

census_file.rename(columns = {'DP05_0005E':'Population under 5 years of age'}, inplace = True)

census_file.rename(columns = {'DP05_0006E':'Population 5 to 9 years old'}, inplace = True)

census_file.rename(columns = {'DP05_0007E':'Population 10 to 14 Years old'}, inplace = True)

census_file.rename(columns = {'DP05_0008E':'Population 15 to 19 Years old'}, inplace = True)

census_file.rename(columns = {'DP05_0009E':'Population 20 to 24 years old'}, inplace = True)

census_file.rename(columns = {'DP05_0010E':'Population 25 to 34 years old'}, inplace = True)

census_file.rename(columns = {'DP05_0011E':'Population 35 to 44 years old'}, inplace = True)

census_file.rename(columns = {'DP05_0012E':'Population 45 to 54 years old'}, inplace = True)

census_file.rename(columns = {'DP05_0013E':'Population 55 to 59 years old'}, inplace = True)

census_file.rename(columns = {'DP05_0014E':'Population 60 to 64 years old'}, inplace = True)

census_file.rename(columns = {'DP05_0015E':'Population 65 to 74 years old'}, inplace = True)

census_file.rename(columns = {'DP05_0016E':'Population 75 to 84 years old'}, inplace = True)

census_file.rename(columns = {'DP05_0017E':'Population 85 years and older'}, inplace = True)

census_file
```
