# TASK-3-PART 2 Enrichment Dataset: Employment data

```python
employment_data = pd.read_csv("../../../../data/stage_1/employment_data.csv")
```

## Displaying employment dataset

```python
employment_data
```

## Reading super dataset from csv file

```python
super_data = pd.read_csv("../../../../data/stage_1/superdataset.csv")
```

## Displaying Super data

```python
super_data
```

### Converting  county_name in employment data to lower case so that it will be easy to map with the superdata.

```python
employment_data['County_Name'] = employment_data['County_Name'].apply(lambda cn: cn.lower().strip())
```

### Splitting the data of county name and state to map with super data

```python
employment_data['County_Name'] = employment_data['County_Name'].apply(lambda cn: cn.split(",")[0])
```

## Displaying employment data

```python
employment_data
```

```python
super_data
```

```python
super_data['County_Name'] = super_data['County_Name'].apply(lambda cn: cn.lower().strip())
super_data['County_Name'] = super_data['County_Name'].apply(lambda cn: cn.split(" ")[0])
```
